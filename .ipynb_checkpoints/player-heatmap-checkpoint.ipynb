```python
#import the needed lib

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mplsoccer.pitch import Pitch
from mplsoccer.pitch import VerticalPitch
import seaborn as sns
```

```python
#call the data using pandas

df = pd.read_csv('./data/player-analyst/qatar-ekuador.csv')
df.head()
```

```python
df.describe()
```

```python
df['X'] = df['X']*1.2
df['Y'] = df['Y']*0.8
df['X2'] = df['X2']*1.2
df['Y2'] = df['Y2']*0.8
```

```python
df.head()
```

## Full Time Movement

```python
#draw the pitch
fig,ax = plt.subplots(figsize=(13.5, 8))
fig.set_facecolor('#22312b')
ax.patch.set_facecolor('#22312b')

pitch = Pitch(pitch_type='statsbomb', half=False, pitch_length=120, pitch_width=80,
             pitch_color='#22312b', line_color='#c7d5cc')
pitch.draw(ax=ax)
plt.gca().invert_yaxis()

#draw the player movements
for x in range(len(df['X'])):
    if df['Player'][x] == "Enner Valencia":
        if df['Mins'][x] <= 45:
            plt.plot([int(df["X"][x]), int(df["X2"][x])], [int(df["Y"][x]), int(df["Y2"][x])], color='green')
            plt.scatter([int(df["X"][x]), int(df["X2"][x])], [int(df["Y"][x]), int(df["Y2"][x])], color="green")
        
        #because the football teams will be change their side in the second half, so we need to flip the X and Y
        elif df['Mins'][x] > 45:
            plt.plot([int(120-df["X"][x]), int(120-df["X2"][x])], [int(80-df["Y"][x]), int(80-df["Y2"][x])], color='blue')
            plt.scatter([int(120-df["X"][x]), int(120-df["X2"][x])], [int(80-df["Y"][x]), int(80-df["Y2"][x])], color="blue")
```

```python
df_player = df[df["Player"] == "M Calceido"]
df_player
```

```python
df_playerCoba = pd.DataFrame(df_player[(df_player["Mins"] >=0) & (df_player["Mins"] <=43)])
df_playerCoba.index = np.arange(0, len(df_playerCoba))
df_playerCoba
```

```python
#draw the pitch
fig,ax = plt.subplots(figsize=(13.5, 8))
fig.set_facecolor('#22312b')
ax.patch.set_facecolor('#22312b')

pitch = Pitch(pitch_type='statsbomb', half=False, pitch_length=120, pitch_width=80,
             pitch_color='#22312b', line_color='#c7d5cc')
pitch.draw(ax=ax)
plt.gca().invert_yaxis()

#draw the player movements
for x in range(len(df_playerCoba['X'])):
    if df_playerCoba['Mins'][x] <= 45:
        plt.plot([int(df_playerCoba["X"][x]), int(df_playerCoba["X2"][x])], [int(df_playerCoba["Y"][x]), int(df_playerCoba["Y2"][x])], color='green')
        plt.scatter([int(df_playerCoba["X"][x]), int(df_playerCoba["X2"][x])], [int(df_playerCoba["Y"][x]), int(df_playerCoba["Y2"][x])], color="green")
    #because the football teams will be change their side in the second half, so we need to flip the X and Y
    elif df_playerCoba['Mins'][x] > 45:
        plt.plot([int(120-df_playerCoba["X"][x]), int(120-df_playerCoba["X2"][x])], [int(80-df_playerCoba["Y"][x]), int(80-df_playerCoba["Y2"][x])], color='blue')
        plt.scatter([int(120-df_playerCoba["X"][x]), int(120-df_playerCoba["X2"][x])], [int(80-df_playerCoba["Y"][x]), int(80-df_playerCoba["Y2"][x])], color="blue")
```

```python
df_player.loc[df_player["Mins"] > 45, "X"] = 120-df_player["X"]
df_player.loc[df_player["Mins"] > 45, "X2"] = 120-df_player["X2"]
df_player.loc[df_player["Mins"] > 45, "Y"] = 80-df_player["Y"]
df_player.loc[df_player["Mins"] > 45, "Y2"] = 80-df_player["Y2"]
```
